```python
### 4. What is the purpose of the code that sets a list of categorical variables
### in a dataset and then casts those variables to the object data type using the astype() function?
# This code converts specific columns in the DataFrame 'data' to the data type 'object' for each item in the list 'catagorialList'.
# Converting the data type to 'object' may be necessary to treat certain columns as categorical variables, enabling easier manipulation,
# analysis, and interpretation of the data.
catagorialList = ['sex','cp','fbs','restecg','exang','ca','thal']
for item in catagorialList:
    data[item] = data[item].astype('object') #casting to object
```

```python
### 6. Explain line 3,4 and 5 and print the shape of x and y

y = data['target'].values
y = y.reshape(y.shape[0],1)
x = data.drop(['target'],axis=1)
##

#y = data['target'].values  take target value
#y = y.reshape(y.shape[0],1)   reshape y array into 1 dimension
#x = data.drop(['target'],axis=1)  drop 'target' in x data
```

```python
### 8. Describe the heart dataset after implementing the min max normalization
#Normalize data (range 0 - 1)
minx = np.min(x)
maxx = np.max(x)
x = (x - minx) / (maxx - minx)
x.head()
```

```python
### 9. Modify the code to split the dataset into train and test (train 70%, val 20% and test 10%).

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=42)
# re-create train and validation set
x_train, x_val, y_train, y_val  = train_test_split(x_train, y_train, test_size=0.2, random_state=42)
# train 70%, validation 20%, test 10%
print(x_train.shape)
print(x_val.shape)
print(x_test.shape)
```

```python
### 10. What is the purpose of each layer in the neural network created using the Sequential() function with 64, 32, and 1 neurons,
### respectively, and softmax and sigmoid activation functions?

model = Sequential() #Allow us to create model layer by layer
model.add(Dense(64, input_dim=21, activation='softmax')) #Softmax turn number data into probabilities which sum to 1
model.add(Dense(32, activation='softmax'))
model.add(Dense(1, activation='sigmoid')) # produce probability value (number between 0 or 1)
model.summary()
```

The first two layers seem to use softmax activation, which is unusual for hidden layers and is typically reserved for the output layer in multiclass classification tasks.
The third layer uses sigmoid activation, which is suitable for binary classification tasks.

```python
### 11. This code compiles a neural network model with a mean squared error loss function, the Adam optimizer with a learning rate of 0.01,
### and accuracy as a performance metric. What does each of these components mean, and how do they affect the model training and performance?

model.compile(loss='mse',
              optimizer=tf.keras.optimizers.Adam(learning_rate=0.01, beta_1=0.9, beta_2=0.999,epsilon=1e-07, amsgrad=False,name='Adam'),
              metrics=['acc'])
```

Mean Squared Error (MSE) Loss Function:

Think of MSE as a measure of how wrong the model's predictions are compared to the actual values.
It calculates the average of the squared differences between the predicted values and the actual values.
Lower MSE means the model's predictions are closer to the actual values.
It's commonly used in regression problems where the goal is to predict continuous values.

Adam Optimizer with Learning Rate of 0.01:

The optimizer is like a driver that steers the model towards better performance.
Adam is an advanced optimization algorithm that adjusts the model's weights to minimize the loss (MSE in this case).
Learning rate (0.01 in this case) determines how big the steps are that the optimizer takes while updating the weights.
A higher learning rate can make the model learn faster but may overshoot the best solution, while a lower learning rate can make the learning process slower but more precise.

Accuracy Metric:

Accuracy is like a report card grade for the model's performance.
It measures the proportion of correct predictions the model makes out of all predictions.
For example, if the model predicts 8 out of 10 cases correctly, its accuracy is 80%.
It's commonly used in classification problems where the goal is to classify inputs into different categories.

How do they affect model training and performance?

MSE loss guides the model to make predictions closer to the actual values, improving its ability to predict continuous outcomes.
Adam optimizer adjusts the model's weights efficiently, helping it learn faster and converge to a better solution.
Accuracy metric gives us a simple understanding of how well the model is performing overall in terms of correct predictions.

**15. Confusion Matrix:**

Definition: A confusion matrix is a performance measurement for classification problems. It is a table that summarizes the performance of a classification model on a set of test data for which the true values are known.

Purpose: It helps visualize the performance of the classification model by presenting a summary of the model's predictions compared to the actual labels. It provides insights into the model's performance in terms of correct and incorrect classifications.
Components:

True Positive (TP): The number of correctly predicted positive instances (i.e., correctly predicted instances of the positive class).

False Positive (FP): The number of incorrectly predicted positive instances (i.e., instances predicted as positive but actually belong to the negative class).

False Negative (FN): The number of incorrectly predicted negative instances (i.e., instances predicted as negative but actually belong to the positive class).

True Negative (TN): The number of correctly predicted negative instances (i.e., correctly predicted instances of the negative class).

**16. Classification Report:**

Definition: A classification report is a text summary of various classification metrics calculated on a classification model's predictions. It provides a comprehensive overview of the model's performance across different classes.
Components:

Precision: The proportion of true positive predictions among all positive predictions made by the model. It measures the model's ability to avoid false positive predictions.

Recall (Sensitivity): The proportion of true positive predictions among all actual positive instances in the dataset. It measures the model's ability to correctly identify positive instances.

F1-Score: The harmonic mean of precision and recall. It provides a balance between precision and recall, especially when the classes are imbalanced.

Support: The number of actual occurrences of each class in the dataset.

Purpose: The classification report helps in evaluating the model's performance in terms of precision, recall, and F1-score for each class. It provides insights into the model's strengths and weaknesses across different classes, facilitating further analysis and model improvement.
